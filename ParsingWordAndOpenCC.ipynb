```python
import jieba
from opencc import OpenCC
```

```python
# Initial
cc = OpenCC('s2t')


# Tokenize
with open('/content/gdrive/MyDrive/專題Project/wiki_text_seg.txt', 'w', encoding='utf-8') as new_f:
    with open('/content/gdrive/MyDrive/專題Project/wiki_text.txt', 'r', encoding='utf-8') as f:
        for times, data in enumerate(f, 1):
            print('data num:', times)
            data = cc.convert(data)
            data = jieba.cut(data)
            data = [word for word in data if word != ' ']
            data = ' '.join(data)

            new_f.write(data)
```
