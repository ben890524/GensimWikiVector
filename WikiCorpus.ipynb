```python
from gensim.corpora import WikiCorpus

wiki_corpus = WikiCorpus('/content/gdrive/MyDrive/專題Project/zhwiki-latest-pages-articles.xml.bz2', dictionary={})
text_num = 0

with open('/content/gdrive/MyDrive/專題Project/wiki_text.txt', 'w', encoding='utf-8') as f:
    for text in wiki_corpus.get_texts():
        f.write(' '.join(text)+'\n')
        text_num += 1
        if text_num % 10000 == 0:
            print('{} articles processed.'.format(text_num))

    print('{} articles processed.'.format(text_num))
```
